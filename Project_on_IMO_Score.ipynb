```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
from matplotlib import style
```

```python
data = pd.read_csv('/content/drive/MyDrive/CERTIFICATE COURSES/My Projects/IMO_EXAM_SCORE.csv')
data.head(15)
```

```python
# Getting the column names
data.columns
```

```python
# storing variables of the data dataframe
years=data['Year']
maths=data['IMO_Maths_Score']
physics=data['IMO_Physics_Score']
chemistry=data['IMO_Chemistry_Score']
computer=data['IMO_Computer_Score']
biology=data['IMO_Biology_Score']
total=data['IMO_Yearwise_Total']
hours=data['Hours_Of_Study_Per_Year']
percentage=data['Percentage_Marks_Received']
```

```python
# Checking the trend of maths and physics marks versus year
plt.plot(years,maths,color='red',marker='s')
plt.plot(years,physics,color='aqua',marker='o')
plt.title('Trend of Maths and Physics Marks over years')
plt.xlabel('Years')
plt.ylabel('Maths and Physics Scores')
plt.legend(['Maths Score','Physics Score'])
plt.show()
```

```python
# Checking the trend of chemistry and biology marks versus year
plt.plot(years,chemistry,color='red',marker='s')
plt.plot(years,biology,color='aqua',marker='o')
plt.title('Trend of Chemistry and Biology Marks over years')
plt.xlabel('Years')
plt.ylabel('Chemistry and Biology Scores')
plt.legend(['Chemistry Score',' Biology Score'])
plt.show()
```

```python
# Checking the trend of computer marks versus year
plt.plot(years,computer,color='gold',marker='s')
plt.title('Trend of Computer Marks over years')
plt.xlabel('Years')
plt.ylabel('Computer Scores')
plt.legend(['Computer Score'])
plt.show()
```

```python
# Plotting all subjects scores in a single plot to understand the respective trend
plt.plot(years,maths,color='red')
plt.plot(years,physics,color='aqua')
plt.plot(years,chemistry,color='gold')
plt.plot(years,computer,color='black')
plt.plot(years,biology,color='lime')
plt.title('Trend of Maths,Physics,Chemistry,Computer and Biology Marks over years')
plt.xlabel('Years')
plt.ylabel('Maths,Physics,Chemistry,Computer and Biology Scores')
plt.legend(['Maths Score','Physics Score','Chemistry Score','Computer Score','Biology Score'])
plt.figure(figsize=(25,25))
plt.show()
```

```python
# Understanding the trend of total marks score versus years
plt.plot(years,total,color='lime')
plt.title('Understanding the trend of total marks in years')
plt.xlabel('Years')
plt.ylabel('Total Marks Score Yearwise')
plt.legend(['Total_Yearwise_Score'])
plt.show()
```

```python
# Understanding the trend of hours study every year versus years
plt.plot(years,hours,color='cyan')
plt.title('Understanding the trend of Hours Study in years')
plt.xlabel('Years')
plt.ylabel('Hours of Study Yearwise')
plt.legend(['Hours of Study Yearwise'])
plt.show()
```

```python
# Observing if there exists any trend in Maths score vs Physics score
plt.plot(maths,physics,'go')
plt.title('Relationship between Maths Score and Physics Score')
plt.xlabel('Maths Score')
plt.ylabel('Physics Score')
plt.legend(['Physics Score'])
plt.show()
```

```python
# Observing if there exists any trend in Maths score vs Chemistry score
plt.plot(maths,chemistry,'ro')
plt.title('Relationship between Maths Score and Chemistry Score')
plt.xlabel('Maths Score')
plt.ylabel('Chemistry Score')
plt.legend(['Chemistry Score'])
plt.show()
```

```python
# Observing if there exists any trend in Maths score vs Computer score
plt.plot(maths,computer,'bo')
plt.title('Relationship between Maths Score and Computer Score')
plt.xlabel('Maths Score')
plt.ylabel('Computer Score')
plt.legend(['Computer Score'])
plt.show()
```

```python
# Observing if there exists any trend in Maths score vs biology score
plt.plot(maths,biology,'bo')
plt.title('Relationship between Maths Score and Biology Score')
plt.xlabel('Maths Score')
plt.ylabel('Biology Score')
plt.legend(['Biology Score'])
plt.show()
```

```python
# Plotting a pie chart on how the numbers affect total score
sizes=[100,95,91,99,96]
labels='Maths','Physics','Chemistry','Computer','Biology'
plt.pie(sizes,labels=labels,autopct='%1.2f%%',colors=['aqua','red','green','purple','gold'],startangle=90)
plt.title('Contribution of Maths, Physics,Chemistry,Computer and Biology in 2017 Marks')
# plt.legend(labels,loc='lower right',title='Subject Name')
plt.figure(figsize=(25,25))
plt.show()
```

```python
# Plotting a donut chart with same data
sizes=[100,95,91,99,96]
labels='MATHS','PHYSICS','CHEMISTRY','COMPUTER','BIOLOGY'
circle=plt.Circle((0,0),0.5,color='white')
plt.pie(sizes,labels=labels,autopct='%1.2f%%',colors=['aqua','red','green','purple','gold'],startangle=90,labeldistance=0.95)
plt.gca().add_artist(circle)
plt.show()
```

```python
# Plotting a bar graph of year vs percentage marks received
plt.barh(years,percentage,color='blue',alpha=0.5)
plt.tight_layout()
plt.xlabel('Percentage of Marks')
plt.ylabel('Years')
plt.title('Distribution pf Percentage of Marks Over Years')
plt.show()
```
